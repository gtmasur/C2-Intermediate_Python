## Histogram
- Histogram gives the idea of distribution of variables (a graph showing frequency distributions).
- In histogram, the data is divided into intervals (bins) and a graph shows the number of observations within each given interval.

# 2. Dictionaries and Pandas (Part 1)
## Dictionaries
- Dictionary is used to store data values in key:value pairs.
- Dictionary items are ordered and do not allow duplicates.
- Dictionary has a unique key element and a value in a dictionary can be again a dictionary.
- Keys have to be immutable (unchangeable) objects. Strings, booleans, integers are immutable objects. A list is mutable so that cannot be used as a key.
- Print keys and values of a dictionary using keys() and values().

## List vs. dictionary
- Both: select, update and remove elements with square brackets [ ]
- List: indexed by range of numbers.
  Dict: indexed by unique keys
- List: collection of values: order matters, when selecting entire subsets. 
  Dict: lookup table with unique keys.

```python
### MAIN CELL
# To see the convenience of dictionary:

# Definition of countries and capital in list format
countries = ['spain', 'france', 'germany', 'norway']
capitals = ['madrid', 'paris', 'berlin', 'oslo']

# Task: print the capital of Germany
ind_ger = countries.index('germany')  # get index of germany
print(capitals[ind_ger])    # print the capital using the corresponding index

# Create a dictionary of countries and capitals
europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin', 'norway':'oslo'}
print(europe['germany'])

# Print out the keys and values of dictionary
print(europe.keys())
print(europe.values())

# Add new data to dict: italy to europe
europe['italy'] = 'rome'
print('italy' in europe)   # checkout italy in europe

#---------------------------------------
# Dictionary manipulation: add and remove data

# Definition of dictionary
europe = {'spain':'madrid', 'france':'paris', 'germany':'bonn',
          'norway':'oslo', 'italy':'rome', 'poland':'warsaw',
          'australia':'vienna' }

# If a key is already available, a value gets updated.
europe['germany'] = 'berlin'   # Update capital of germany
# Remove data from dict
del(europe['australia'])


# Dictionary of dictionaries
europe = { 'spain': { 'capital':'madrid', 'population':46.77 },
           'france': { 'capital':'paris', 'population':66.03 },
           'germany': { 'capital':'berlin', 'population':80.62 },
           'norway': { 'capital':'oslo', 'population':5.084 } }

# Print out the capital of France
print(europe['france']['capital'])

# Add data after creating sub-dictionary data
data = {'capital':'rome', 'population':59.83}
europe['italy'] = data
print(europe)
```

# 2. Dictionaries and Pandas (Part 2)
## Pandas
- Pandas provides easy-to-use data structures, built on Numpy, is high level data manipulation tool.
- DataFrame is the most important data structure of Panda.
- In DataFrame, rows represent the observation and columns represent variables.
- Possible to add unique row labels in DataFrame.
- Ways to create DataFrame:
  1. From dict: keys are columns labels and values are data, column by column.  
  2. From CSV file: (CSV: comma separated values)
- Creating a DataFrame from a dictionary works, however it is not very efficient. In the case of a big data with a regular structure, such as CSV files, create it from this file.


## DataFrame : Data selection
- Pandas Series are 1D labelled array. A bunch of series together create a DataFrame.
- Data Selection through square brackets:
  1. Column access:
    - dframe_name[column_label]: selected data is a Series
    - dframe_name[[column_label]]: selected data is a DataFrame
    - Row access: only through slicing as in list and array, dframe_name[1:4]
  2. One cannot choose row and columns at the same time.
- When an extended functionality is required similar to 2D Numpy Arrays, there are
  1. loc (label- based)
  2. iloc (integer position-based)
- As above, pandas Series and pandas DataFrame apply here too through square brackets.
- loc and iloc functions allows rows and column access simultaneously.
- loc is label-based, which means that one has to specify rows and columns based on their row and column labels. 
- iloc is integer index based, so one has to specify rows and columns by their integer index.

```python
### MAIN CELL
#-------------------------------------
# Pandas DataFrame

# 1) Create a DataFrame from a dict

import pandas as pd

names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]   # the direction of driving cars is right
cpc = [809, 731, 588, 18, 200, 70, 45]   # the number of motor vehicles per 1000 people

# Create a dict by putting above lists as values
cars_dict = { 'cars_per_cap':cpc, 'country':names, 'drives_right':dr}
# Create DataFrame from a dict.
cars = pd.DataFrame(cars_dict)
# Specify row labels for countries
cars.index = ['US', 'AUS', 'JAP', 'IN', 'RU', 'MOR', 'EG']
print('Writing a DataFrame from a dictionary:')
print(cars)

#-------------------------------------
# 2) Create a DataFrame from a CSV file

# Get rid of the first indexing column by index_col = 0.
#cars = pd.read_csv('cars.csv')
cars = pd.read_csv('cars.csv', index_col=0)  
print('\n'+'Writing a DataFrame from a CSV file:')
print(cars)

#======================================
# Data Selection

# 1) Square brackets:

# Print out country column 
print('\nWriting data as pandas Series:')
print(cars['country'])    # as Pandas Series
print('\nWriting data as pandas DataFrame:')
print(cars[['country']])  # as Pandas DataFrame
print('\nWriting two columns of data:')
print(cars[['country', 'drives_right']])
print('\nWriting the selected rows of data:')
print(cars[3:6])  # fourth, fifth and sixth observations


#-------------------------------------
# 2) loc and iloc functions

print("\nPrinting data using loc and iloc:")

print("\nPrinting the row selection of data:")
# Print out observation for Japan in Series
print(cars.loc['JAP'])
print(cars.iloc[2])
# Print out observations for Australia and Egypt in DF
print(cars.loc[['AUS', 'EG']])
print(cars.iloc[[1,-1]])

print("\nPriting the column selection of data:")
# Print out drives_right column as Series
print(cars.loc[:,'drives_right'])   # [rows, columns]
print(cars.iloc[:,2])
# Print out drives_right column as DataFrame
print(cars.loc[:,['drives_right']])
print(cars.iloc[:,[2]])
#print(type(cars.iloc[:,2]), type(cars.iloc[:,[2]]))

print("\nPrinting the more than one column selections of data:")
# Print out cars_per_cap and drives_right as DataFrame
print(cars.loc[:,['cars_per_cap','drives_right']])
print(cars.iloc[:,[0,2]])

print("\nPrinting the row and columns selection of data:")
# Print out drives_right value of Morocco
print(cars.loc['MOR', 'drives_right'])
print(cars.iloc[5,2])
# Print sub-DataFrame
print(cars.loc[['RU', 'MOR'], ['country', 'drives_right']])
print(cars.iloc[[4,5], [1, 2]])
```

```python
### MAIN CELL
# Array equivalence of booleans operators

import numpy as np

# sizes of rooms in arrays
my_house = np.array([18.0, 20.0, 10.75, 9.50])
your_house = np.array([14.0, 24.0, 14.25, 9.0])

# my_house greater than 18.5 or smaller than 10
print(np.logical_or(my_house >18.5, my_house<10))

# Both my_house and your_house smaller than 11
print(np.logical_and(my_house<11, your_house<11))

#====================================================
# Filtering pandas DataFrame

import pandas as pd

cars = pd.read_csv('cars.csv', index_col = 0)

# Extract drives_right column as Series and use dr to subset cars 
dr = cars['drives_right'] 
sel = cars[dr]
# sel = cars[cars['drives_right']]  # in a one-linear line
print(sel)

#--------------------------------
# Create car_maniac: observations that have a cars_per_cap over 500
cpc = cars['cars_per_cap']
many_cars = cpc > 500
car_maniac = cars[many_cars]
print(car_maniac)

#-------------------------
# Create medium: observations with cars_per_cap between 100 and 500
cpc = cars['cars_per_cap']
between= np.logical_and(cpc >100, cpc < 500)
medium = cars[between]
print(medium)
```

## for loop
- for < variable > in sequence:   expression

- for index, < variable > in enumerate(sequence): 
    
## Loops data structures

  - for key, value in dict.items():  expression

  - 1D arrays: for val in array: expression (the same)
  - 2D arrays: for val in np.nditer(array): expression

- 3) Pandas DataFrame: 
   - every observation is iterated over and on every iteration the row label and actual row contents are available:
   - for label, row in dframe.iterrows(): expression
   - label: row label, row: row data as pandas Series
   - possible to vectorize it using apply() method.

```python
### MAIN CELL

# while loop

# Code a while loop that keeps running as long as offset is not equal to 0
# If offset is greater than zero, you should decrease offset by 1.
# Else, you should increase offset by 1.  
offset = 8
while offset != 0 :
    print("correcting...")
    if offset > 0 :
      offset = offset - 1
    else : 
      offset = offset + 1
    print(offset)    
    
#---------------------------------------------
# for loop for lists

# np_height: a NumPy array containing the heights of Major League Baseball players, 
# np_baseball: a 2D NumPy array containing the heights (first column) and weights (second column).

np_height = [74, 74, 72, 72, 73, 69, 69, 71, 76, 71, 73, 73, 74, 74, 69, 70, 73, 75, 78, 79]
np_baseball = [[74, 180],[74, 215], [72, 210], [72, 210], [73, 188], [69, 176], [69, 209], [71, 200], 
               [76, 231], [71, 180], [73, 188], [73, 180], [74, 185], [74, 160], [69, 180], [70, 185], 
               [73, 189], [75, 185], [78, 219], [79, 230]]

# For loop over np_height
for x in np_height:  print(str(x) + ' inches')
for x,y in enumerate(np_height):
    print('The player'+ str(x) + ' has ' + str(y)+' inches height.')

# For loop for list of lists
for item in np_baseball: 
    print('height: '+ str(item[0]) + ' weight: ' + str(item[1]))
# See the difference between the two followings:
for x in np_baseball: print(x)               # prints items
for x in np.nditer(np_baseball):  print(x)   # prints each entries

#---------------------------------------------
# for loop for dictionary

europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin',
          'norway':'oslo', 'italy':'rome', 'poland':'warsaw', 'austria':'vienna' }
          
# Iterate over europe
for key, value in europe.items():
    print('the capital of '+ key + ' is '+ value)

#-------------------------------------------------
# For loop for DataFrames

import pandas as pd

# Import cars data
cars = pd.read_csv('cars.csv', index_col = 0)

# Iterate over rows of cars
for lab, row in cars.iterrows():
    print(lab)  # row label
    print(row)  # row content

# easily select variables from the Pandas Series using square brackets:
# Adapt for loop to print "country: cars_per_cap"
for lab, row in cars.iterrows() :
    print( lab + ': ' + str(row['cars_per_cap']) )    
      
# Code for loop to add a new column to DataFrame
for lab, row in cars.iterrows():
    cars.loc[lab, 'country'.upper()] = row['country'].upper()
    
# In every iteration of for loop, a new pandas Series is created.    
# instead, use .apply(str.upper) to vectorize it.
cars['country'.upper()] = cars['country'].apply(str.upper)     
```

# 5. Case Study: Hacker Statistics
## Random numbers
- Hacker statistics: simulating a process a thousands of times to see in what fraction of the simulations that the process occurs.
-  Random generators:
   - np.random.rand()
   - np.random.seed(123), np.random.rand(): create from a seed, same seed=same random numbers
   - np.random.randint(start,end): random integer between the given interval
## Random Walk
- random step: using a dice to determine the next step
- random walk: using a dice 100 times to determine the next step, a succession of random steps
## Distribution
- distribution of random walks
- To calculates chances, one needs to repeatedly simulate the random walk and collect the results.
- Each random walk will end with different final step.
- If one simulates the process 10000 times, one receives 10000 of final steps, which is the distribution of final steps
- After knowing dist., we cna calculate chances.

```python
### MAIN CELL

# i) Random numbers

# Import numpy as np
import numpy as np

# Set the seed
np.random.seed(123)

# Generate and print random float
print(np.random.rand())

# Use randint() to simulate a dice
print(np.random.randint(1,7))

#---------------------------------

# Problem to solve:
# If dice is 1 or 2, you go one step down.
# if dice is 3, 4 or 5, you go one step up.
# Else, you throw the dice again. The number on the dice is the number of steps you go up.

# NumPy is imported, seed is set

# Starting step
step = 50

# Roll the dice
dice = np.random.randint(1,7)

# Finish the control construct
if dice <= 2 :
    step = step - 1
elif dice <= 5:
    step = step + 1
else :
    step = step + np.random.randint(1,7)

# Print out dice and step
print(dice,step)

#--------------------------------------
# ii) Random walk

# Initialize random_walk
random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)

    if dice <= 2:
        # use max to make sure step can't go below 0
        step = max(0,step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)
    random_walk.append(step)

import matplotlib.pyplot as plt
plt.plot(random_walk)
plt.show()

#-------------------------------------
# iii) Distribution

# Question: what is the chance that one reaches 60 steps high?

# numpy and matplotlib imported, seed set.

# initialize and populate all_walks
all_walks = []
for i in range(5) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to NumPy array: np_aw
np_aw = np.array(all_walks)
#print(np_aw.shape)
#print(np_aw)
# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)
# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()

#---------------------------------------

# numpy and matplotlib imported, seed set

# clear the plot so it doesn't get cluttered if you run this many times
plt.clf()

# Simulate random walk 20 times
all_walks = []
for i in range(20) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        # You're a bit clumsy and you have a 0.5% chance of falling down. 
        if np.random.rand() < 0.005 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()

#----------------------------------
# you want to know about the end points of all the random walks you've simulated.
# These end points have a certain distribution that you can visualize with a histogram.
# TODO: set above first for loop to 500. simulate the random walk 500 times

# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]
print(ends.shape, np_aw_t.shape)

# Plot histogram of ends, display plot
plt.hist(ends, bins = 10)
plt.show()

#----------------------------
# To calculate the chance that this end point is greater than or equal to 60,
# you can count the number of integers in ends that are greater than or equal 
# to 60 and divide that number by 500, the total number of simulations.

count = 0
for x in range(len(ends)):
    if x >= 60:
        count = count +1
print(count) 
print(count/500)
```
